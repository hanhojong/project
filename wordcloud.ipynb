```python
# 데이터 전처리
text = ''

with open("C:/Users/gksgh/OneDrive/바탕 화면/워드클라우드예시.txt", "r", encoding="utf-8") as f:
    lines = f.readlines()
    for line in lines[4:] :
        if '] [' in line :
            text += (line.split('] ')[2].replace('이모티콘', "").replace('!', "").replace('?', "").replace('ㅎ', "")
            .replace("사진", "").replace('삭제된 메세지입니다.', "").replace('ㅇ', '').replace("샵검색", '').replace('ㅋ', "")
            .replace("txt", "").replace("pptx", "").replace("웅웅", "").replace("응응", "").replace("파일", "").replace("나", "")
            .replace("ㄱ\n", "").replace("융융", "").replace("음", "").replace("아", "").replace("이미지", "").replace("고\n", "")
            .replace("내가", "").replace("https", "").replace("이거", "").replace("좀", "").replace("노", "").replace("hwp", "")
            .replace("오케이", ""))
print(text)
```

```python
# 폰트 지정 및 파일경로 설정
import matplotlib.font_manager as fm
for font in fm.fontManager.ttflist :
    if 'Gothic' in font.name :
        print(font.name, font.fname)
font_path = 'C:/Windows/Fonts/Hancom Gothic Bold.ttf'
```

```python
# 워드클라우드 배경 지정
from PIL import Image
import numpy as np
mask = np.array(Image.open('C:/Users/gksgh/OneDrive/바탕 화면/love.png'))
```

```python
# 워드클라우드 도출
from wordcloud import WordCloud
wc = WordCloud(font_path=font_path, background_color="white", colormap="Blues", mask = mask,prefer_horizontal = True)
wc.generate(text)
wc.to_file("example.png")
```
